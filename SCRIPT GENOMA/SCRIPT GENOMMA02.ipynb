```python
import os

def listar_arquivos_excel(diretorio):
    """
    Retorna todos os arquivos .xlsx ou .xls de um diretório.

    Parâmetros:
    diretorio (str): O caminho do diretório.

    Retorna:
    lista_arquivos (list): Uma lista de arquivos .xlsx ou .xls no diretório.
    """
    lista_arquivos = []
    
    try:
        # Iterar sobre os arquivos no diretório
        for arquivo in os.listdir(diretorio):
            # Verificar se o arquivo termina com .xlsx ou .xls
            if arquivo.endswith('.xlsx') or arquivo.endswith('.xls'):
                lista_arquivos.append(arquivo)
        
        return lista_arquivos
    except Exception as e:
        print(f"Erro ao listar arquivos no diretório: {e}")
        return []

# Exemplo de uso da função
def mover_arquivo(caminho_origem, caminho_destino):
    """
    Move um arquivo de uma pasta para outra.

    Parâmetros:
    caminho_origem (str): O caminho completo do arquivo de origem.
    caminho_destino (str): O caminho completo do arquivo de destino.
    """
    try:
        # Verifica se o arquivo de origem existe
        if not os.path.exists(caminho_origem):
            raise FileNotFoundError(f"Arquivo de origem não encontrado: {caminho_origem}")

        # Cria o diretório de destino se ele não existir
        os.makedirs(os.path.dirname(caminho_destino), exist_ok=True)
        
        # Move o arquivo
        shutil.move(caminho_origem, caminho_destino)
        print(f"Arquivo movido com sucesso de {caminho_origem} para {caminho_destino}")
    except Exception as e:
        print(f"Erro ao mover o arquivo: {e}")

```
